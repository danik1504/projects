```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = car_df.drop(['Customer Name','Customer e-mail', 'Country', 'Car Purchase Amount'], axis = 1)
```

```python
y = car_df['Car Purchase Amount']
```

```python
X
```

```python
y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
scaler.data_max_
```

```python
y = y.values.reshape(-1,1)
```

```python
y_scaled = scaler.fit_transform(y)
```

```python
y_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.25, random_state=101)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(25, input_dim = 5, activation = 'relu'))
model.add(Dense(25, activation ='relu'))
model.add(Dense(1, activation = 'linear'))
```

```python
model.summary()
```

```python
 model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python

epochs_hist = model.fit(X_train, y_train, epochs=20, batch_size=50,  verbose=1, validation_split=0.2)
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss Progress During Training')
plt.ylabel('Training and Validation Loss')
plt.xlabel('Epoch number')
plt.legend(['Training Loss', 'Validation Loss']);
```

```python
X_test = np.array([[1, 50, 50000, 10000, 600000]])
```

```python
y_predict = model.predict(X_test)
```

```python
print("Expected Purchase Amount", y_predict)
```
